```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Asia'

page = requests.get(url)
    
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
table = soup.find_all('table')[1]
```

```python
asia_titles = table.find_all('th', scope = 'col', limit=7)

asia_titles
```

```python
asia_table_titles = [title.text.strip() for title in asia_titles]

print(asia_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = asia_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[2:]:
    row_data = row.find_all('th') + row.find_all('td')
    c_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = c_row_data
```

```python
df
```
